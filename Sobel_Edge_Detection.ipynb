```python
from PIL import Image
```

```python
!wget https://www.formula1.com/content/dam/fom-website/2018-redesign-assets/Racehub%20header%20images%2016x9/Vietnam.jpg.transform/9col/image.jpg
```

```python
im=Image.open('image.jpg')
```

```python
import numpy as np
gray=np.mean(im,axis=2)
```

```python
hx=np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
hy=np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
```

```python
from scipy.signal import convolve2d
```

```python
o1=convolve2d(hx,gray)
```

```python
o2=convolve2d(hy,gray)
```

```python
o=np.sqrt(np.square(o1)+np.square(o2))
```

```python
import matplotlib.pyplot as plt
plt.imshow(o,cmap='gray')
```
